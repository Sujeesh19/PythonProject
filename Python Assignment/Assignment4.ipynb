```python
# write the function which has to show 1 to 100 numbers.
def printNumber():
    for i in range(1, 101):
        print(i, end=" ")
printNumber()
```

```python
# take any 4 random def function example(the code should contain atleast 6 lines).
def Calculator(x, y):
    a = x+y
    b = x-y
    c = x*y
    d = x/y
    print(a,b,c,d)
x = int(input("Enter the First Number: "))
y = int(input("Enter the Second Number: "))
Calculator(x, y)
```

```python
def fibo(num):
    n1, n2 = 0, 1
    count = 0
    while count < num:
       print(n1, end=" ")
       nth = n1 + n2
       n1 = n2
       n2 = nth
       count += 1
fibo(7)
```

```python
def OddorEven(num):
    if num%2 == 0:
        print(num, " is an even number")
    else:
        print(num, " is a odd number")
num = int(input("Enter the number: "))
OddorEven(num)
```

```python
def inputCheck(n):
    if type(n) == int:
        print(n, " is an Integer")
    if type(n) == str:
        print(n, " is a String")
    if type(n) == float:
        print(n, " is a float")
inputCheck(1)
inputCheck("hello")
inputCheck(7.5)
```

Q. why we use exception handling in python?
* An exception is an event, which occurs during the execution of a program that disrupts the normal flow of the program's instructions. So we use exception handling for error handling so that code functions normally.

Q. what us the use of exception handling?
* Improved program reliability
* Simplified error handling
* Easier Debugging

Q. what is disadvantage of exception handling?
* Increases Code Complexity
* Increases Space Complexity
* Increases Time Complexity
* Security risks

```python
# take any 3 examples of exception handling (which contain try, except, else, finally, keyword)
try:
    a = 10
    b = "Sujeesh"
    c = a+b
except TypeError:
    print("Check your code again")
else: 
    print("executed successfully")
finally:
    print("This is my execution")
```

```python
try:
    a = int(input("Enter first number: "))
    b = int(input("Enter second number: "))
    c= a/b
except ZeroDivisionError:
    print("Denominator should not be zero")
else:
    print(c)
finally:
    print("This is my execution")
```

```python
try: 
    a = int(input("Enter any number "))
except ValueError:
    print("Enter Integer value")
else:
    print(a)
finally:
    print("This is my execution")
```
