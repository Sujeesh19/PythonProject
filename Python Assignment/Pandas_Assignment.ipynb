```python
# First step is to importing the library
import pandas as pd
```

```python
# Second Step to read the dataset
df = pd.read_csv("FlipkartMoblies.csv")
```

```python
# Printing the dataset
df
```

```python
# Checking the shape of the dataset
df.shape
```

```python
# prints information about the DataFrame
df.info()   
```

```python
# Decribe function gives description of the dataset
df.describe()
```

```python
# Checking for how many null values are present
df.isnull().sum()
```

```python
# Filling Rating column null values with the mean value
df['Rating'].fillna(value=df['Rating'].mean(), inplace = True)
```

```python
# Again Checking for how many null values are present
df.isnull().sum()
```

```python
# Filling the Memory column null values with the mode value
df['Memory'].fillna(value=df['Memory'].mode()[0], inplace=True)
```

```python
# Again Checking for how many null values are present
df.isnull().sum()
```

```python
# Filling the Storage column null values with the mode value
df['Storage'].fillna(value=df['Storage'].mode()[0], inplace=True)
```

```python
# Again Checking for how many null values are present
df.isnull().sum()
```

```python
# Checking for duplicates count
df.duplicated().sum()
```

```python
# Removing the duplicate rows from the dataset
df.drop_duplicates(inplace = True)
```

```python
# Checking for is their any duplicate rows left
df.duplicated().sum()
```
